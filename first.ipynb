```python
def plot_attrition_by_category(df, column, chart="bar", sort=True):
    """
    특정 범주형 변수별 이직률 시각화
    column : 범주형 변수 이름
    chart : "bar" or "pie"
    sort : True면 이직률 기준 정렬
    """
    # 이직 여부 플래그
    if "Attrition_Flag" not in df.columns:
        df["Attrition_Flag"] = df["Attrition"].map({"Yes": 1, "No": 0})

    # 그룹별 평균 이직률 계산
    attrition_rate = df.groupby(column)["Attrition_Flag"].mean() * 100

    # 정렬 옵션 적용
    if sort:
        attrition_rate = attrition_rate.sort_values(ascending=False)

    # 출력
    print(f"\n[{column}] 별 이직률 (%)")
    print(attrition_rate.round(2))

    # 시각화
    plt.figure(figsize=(8,6))
    if chart == "bar":
        sns.barplot(
            x=attrition_rate.index,
            y=attrition_rate.values,
            palette="Set2"
        )
        plt.ylabel("Attrition Rate (%)")
        plt.title(f"{column} 별 이직률 (%)")
        plt.xticks(rotation=45)
    elif chart == "pie":
        plt.pie(
            attrition_rate,
            labels=[f"{cat} ({rate:.1f}%)" for cat, rate in attrition_rate.items()],
            autopct='%1.1f%%',
            startangle=90
        )
        plt.title(f"{column} 별 이직률 (%)")
    plt.show()
```

```python
# 결혼 상태별 이직률
plot_attrition_by_category(df, "MaritalStatus", chart="pie")

# 직무별 이직률
plot_attrition_by_category(df, "JobRole", chart="bar")

# 출장 빈도별 이직률
plot_attrition_by_category(df, "BusinessTravel", chart="pie")

# 전공별 이직률
plot_attrition_by_category(df, "EducationField", chart="bar")

# 직급별 이직률
plot_attrition_by_category(df, "JobLevel", chart="bar")

# 성과 평가별 이직률
plot_attrition_by_category(df, "PerformanceRating", chart="bar")

# 일과 삶의 균형별 이직률
plot_attrition_by_category(df, "WorkLifeBalance", chart="pie")

# 직무 만족도별 이직률
plot_attrition_by_category(df, "JobSatisfaction", chart="bar")

# 근무 환경 만족도별 이직률
plot_attrition_by_category(df, "EnvironmentSatisfaction", chart="bar")

# 대인 관계 만족도별 이직률
plot_attrition_by_category(df, "RelationshipSatisfaction", chart="bar")
```

```python
def plot_attrition_pie(df, column, title=""):
    """
    Sales 데이터 내 특정 범주형 변수별 이직률(%)을 파이차트로 시각화
    """
    # 이직 여부 플래그
    if "Attrition_Flag" not in df.columns:
        df["Attrition_Flag"] = df["Attrition"].map({"Yes": 1, "No": 0})
    
    # 그룹별 이직률 계산
    attrition_rate = df.groupby(column)["Attrition_Flag"].mean() * 100
    print(f"\n[{column}] 별 이직률 (%)")
    print(attrition_rate.round(2))
    
    # 파이차트 시각화
    plt.figure(figsize=(6,6))
    plt.pie(
        attrition_rate,
        labels=[f"{cat} ({rate:.1f}%)" for cat, rate in attrition_rate.items()],
        autopct='%1.1f%%',
        startangle=90,
        colors=sns.color_palette("Set2", len(attrition_rate))
    )
    plt.title(title if title else f"{column}별 이직률")
    plt.show()
```

```python
# Sales 부서만 필터링
df_sales = df[df["Department"] == "Sales"].copy()

# 성별별 이직률
plot_attrition_pie(df_sales, "Gender", "Sales 부서 성별별 이직률")

# 결혼 상태별 이직률
plot_attrition_pie(df_sales, "MaritalStatus", "Sales 부서 결혼 상태별 이직률")

# 출장 빈도별 이직률
plot_attrition_pie(df_sales, "BusinessTravel", "Sales 부서 출장 빈도별 이직률")

# 직무 만족도별 이직률
plot_attrition_pie(df_sales, "JobSatisfaction", "Sales 부서 직무 만족도별 이직률")

# 직무 몰입도별 이직률
plot_attrition_pie(df_sales, "JobInvolvement", "Sales 부서 직무 몰입도별 이직률")
```

```python
import statsmodels.api as sm
import pandas as pd

# 타겟 변수
df["Attrition_Flag"] = df["Attrition"].map({"Yes":1, "No":0})

# 설명 변수 (급여, 직급 통제 + 관심 변수 포함)
X = df[["MonthlyIncome","JobLevel","Age","MaritalStatus","Department","OverTime"]]
y = df["Attrition_Flag"]

# 더미 처리
X = pd.get_dummies(X, drop_first=True)

# 결측치 제거 + 숫자형 변환
X = X.dropna().astype(float)
y = y.loc[X.index].astype(float)

# 로지스틱 회귀
model = sm.Logit(y, sm.add_constant(X)).fit()
print(model.summary())
```

🔹 주요 결과 해석

OverTime_Yes (초과근무 있음, coef=1.48, p<0.001)

초과근무하는 직원은 이직할 확률이 exp(1.48) ≈ 4.4배 높음

✅ 이직 방지 핵심 포인트: 초과근무 줄이기, 워라밸 정책 강화

MaritalStatus_Single (미혼, coef=1.13, p<0.001)

미혼 직원은 기혼 대비 이직 확률이 exp(1.13) ≈ 3.1배 높음

✅ 전략: 미혼 직원에게 커리어 성장 기회, 멘토링, 자기계발 지원 제공

Age (나이, coef=-0.024, p=0.018)

나이가 많을수록 이직 확률이 감소 (coef 음수)

✅ 젊은 직원(20–30대) 이직 위험 ↑ → 초기 경력 개발 프로그램 필요

JobLevel (직급, coef=-0.47, p=0.047)

직급이 높을수록 이직 확률 감소

✅ 낮은 직급(Entry-level) 직원 유지 전략 필요 → 승진/보상 트랙 제시

MonthlyIncome (월급, p=0.882 → 유의하지 않음)

급여 자체는 통제했을 때 이직률에 유의미한 영향 없음

즉, 급여보다 다른 요인(초과근무, 만족도 등)이 더 중요

부서 효과

Sales / R&D / HR 비교했을 때 부서 자체는 유의미하지 않음 (p>0.05)

이직 차이는 부서 특성보다는 초과근무, 직무/만족도 같은 요인에서 발생

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 연령대 구간 정의
bins = [18, 29, 39, 49, 60]
labels = ["20대 이하","30대","40대","50대 이상"]
df["AgeGroup"] = pd.cut(df["Age"], bins=bins, labels=labels, right=True)

# 연령대별 초과근무 비율 계산
overtime_rate = df.groupby("AgeGroup")["OverTime"].apply(lambda x: (x=="Yes").mean() * 100)

print("연령대별 초과근무 비율 (%)")
print(overtime_rate.round(2))

# 시각화 (막대그래프)
plt.figure(figsize=(6,4))
sns.barplot(x=overtime_rate.index, y=overtime_rate.values, palette="Set2")
plt.title("연령대별 초과근무 비율 (%)")
plt.ylabel("OverTime Rate (%)")
plt.xlabel("Age Group")
plt.show()

# 시각화 (파이차트, 원하면)
plt.figure(figsize=(6,6))
plt.pie(
    overtime_rate,
    labels=[f"{grp} ({rate:.1f}%)" for grp, rate in overtime_rate.items()],
    autopct="%1.1f%%",
    startangle=90,
    colors=sns.color_palette("pastel", len(overtime_rate))
)
plt.title("연령대별 초과근무 비율")
plt.show()
```

```python
import statsmodels.api as sm
import pandas as pd

# 독립 변수들 (ID/상수 제외)
features = [
    "Age","BusinessTravel","DailyRate","Department","DistanceFromHome","Education",
    "EducationField","Gender","HourlyRate","JobInvolvement","JobLevel","JobRole",
    "MaritalStatus","MonthlyIncome","MonthlyRate","NumCompaniesWorked","OverTime",
    "PercentSalaryHike","PerformanceRating","RelationshipSatisfaction","StockOptionLevel",
    "TotalWorkingYears","TrainingTimesLastYear","YearsAtCompany","YearsInCurrentRole",
    "YearsSinceLastPromotion","YearsWithCurrManager"
]

X = pd.get_dummies(df[features], drop_first=True).astype(float)

# -----------------------------
# 1) JobSatisfaction (3~4 vs 1~2)
# -----------------------------
df["JobSat_Flag"] = (df["JobSatisfaction"] >= 3).astype(int)
y1 = df["JobSat_Flag"]

model1 = sm.Logit(y1, sm.add_constant(X)).fit()
print("\n[JobSatisfaction] Logistic Regression Results")
print(model1.summary())


```

```python
summary_table = pd.DataFrame({
    "coef": model1.params,
    "std_err": model1.bse,
    "z": model1.tvalues,
    "p_value": model1.pvalues,
    "CI_lower": model1.conf_int()[0],
    "CI_upper": model1.conf_int()[1]
})

summary_table = summary_table.sort_values(by="p_value")

print(summary_table)
```

```python
# -----------------------------
# 2) WorkLifeBalance (3~4 vs 1~2)
# -----------------------------
df["WLB_Flag"] = (df["WorkLifeBalance"] >= 3).astype(int)
y2 = df["WLB_Flag"]

model2 = sm.Logit(y2, sm.add_constant(X)).fit()
print("\n[WorkLifeBalance] Logistic Regression Results")
print(model2.summary())
```

```python
# WorkLifeBalance (3~4 vs 1~2)
df["WLB_Flag"] = (df["WorkLifeBalance"] >= 3).astype(int)
y2 = df["WLB_Flag"]

model2 = sm.Logit(y2, sm.add_constant(X)).fit()

# 결과 테이블 변환
summary_table2 = pd.DataFrame({
    "coef": model2.params,
    "std_err": model2.bse,
    "z": model2.tvalues,
    "p_value": model2.pvalues,
    "CI_lower": model2.conf_int()[0],
    "CI_upper": model2.conf_int()[1]
})

# p-value 기준 정렬
summary_table2 = summary_table2.sort_values(by="p_value")

print("\n[WorkLifeBalance] Logistic Regression Results (정리)")
print(summary_table2)

# p < 0.05 필터링 (유의미한 변수만)
significant_wlb = summary_table2[summary_table2["p_value"] < 0.05]
print("\n[WorkLifeBalance] 유의미한 변수 (p < 0.05)")
print(significant_wlb)
```
