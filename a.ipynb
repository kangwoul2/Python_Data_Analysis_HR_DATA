```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rc('font', family='AppleGothic')

# 1. CSV 읽기
df = pd.read_csv("jobrole_workload_summary.csv", index_col=0)

# 시각화 스타일
sns.set(style="whitegrid")
plt.rcParams.update({"figure.figsize": (10,6)})

# ---------------------------------------------------
# 2. Total_Workload 큰 순서로 막대그래프
plt.figure()
df_sorted_total = df.sort_values("Total_Workload", ascending=False)
sns.barplot(x=df_sorted_total["Total_Workload"],
            y=df_sorted_total.index,
            palette="Blues_r")
plt.title("Total Workload by JobRole (Descending)")
plt.xlabel("Total Workload")
plt.ylabel("JobRole")
plt.tight_layout()
plt.show()

# ---------------------------------------------------
# 3. Extra_Cost 적은 순서로 막대그래프
plt.figure()
df_sorted_cost = df.sort_values("Extra_Cost", ascending=True)
sns.barplot(x=df_sorted_cost["Extra_Cost"],
            y=df_sorted_cost.index,
            palette="Greens")
plt.title("Extra Cost by JobRole (Ascending)")
plt.xlabel("Extra Cost")
plt.ylabel("JobRole")
plt.tight_layout()
plt.show()

# ---------------------------------------------------
# 4. 각 변수별 값 크기 순서로 나열
# (직무별이 아닌, 열별 내림차순 리스트 출력)
for col in df.columns:
    print(f"\n=== {col} 내림차순 ===")
    print(df[col].sort_values(ascending=False))
```

```python
import pandas as pd

# 1. CSV 파일 읽기
salary = pd.read_csv("department_jobrole_salary_summary.csv")
workload = pd.read_csv("department_jobrole_workload_summary.csv")

# workload 파일이 MultiIndex(Department, JobRole)라면 reset_index로 풀기
if isinstance(workload.index, pd.MultiIndex):
    workload = workload.reset_index()

# 2. Department, JobRole 기준으로 병합
# how='outer' 를 사용하면 양쪽 파일에 없는 행도 모두 유지
merged = pd.merge(
    salary,
    workload,
    on=["Department", "JobRole"],
    how="outer"
)

# 3. CSV 저장
merged.to_csv("total.csv", index=False, encoding="utf-8-sig")

print("통합 파일 저장 완료: total.csv")
print(merged.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ---------------------------------------------------
# 1. 데이터 불러오기
df = pd.read_csv("total.csv", encoding="utf-8-sig")

# 2. 파생 변수 계산
df["평균_생산량_1인당"] = df["Total_Workload"] / df["총인원수"]
df["추가_고용비용"] = df["Extra_People_Needed"] * df["평균월급"]

# 부서+직무를 하나의 문자열로 묶어 x축에 사용
df["부서_직무"] = df["Department"] + " - " + df["JobRole"]

# ---------------------------------------------------
sns.set(style="whitegrid")
plt.figure(figsize=(12,6))
# 3-1. 1인당 평균 생산량 (내림차순)
order1 = df.sort_values("평균_생산량_1인당", ascending=False)
sns.barplot(data=order1, x="평균_생산량_1인당", y="부서_직무", palette="Blues_r")
plt.title("부서+직무별 1인당 평균 생산량")
plt.xlabel("1인당 평균 생산량")
plt.ylabel("부서 - 직무")
plt.tight_layout()
plt.show()

# ---------------------------------------------------
plt.figure(figsize=(12,6))
# 3-2. 추가 고용해야 할 인원
order2 = df.sort_values("Extra_People_Needed", ascending=False)
sns.barplot(data=order2, x="Extra_People_Needed", y="부서_직무", palette="Oranges_r")
plt.title("부서+직무별 추가 고용 필요 인원")
plt.xlabel("추가 고용 필요 인원")
plt.ylabel("부서 - 직무")
plt.tight_layout()
plt.show()

# ---------------------------------------------------
plt.figure(figsize=(12,6))
# 3-3. 퇴직률 감소를 위해 추가 고용 시 필요한 비용
order3 = df.sort_values("추가_고용비용", ascending=False)
sns.barplot(data=order3, x="추가_고용비용", y="부서_직무", palette="Greens_r")
plt.title("부서+직무별 퇴직률 완화를 위한 추가 고용 비용")
plt.xlabel("추가 고용 비용 (월 단위)")
plt.ylabel("부서 - 직무")
plt.tight_layout()
plt.show()
```
