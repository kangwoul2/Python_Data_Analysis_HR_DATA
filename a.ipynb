```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rc('font', family='AppleGothic')

# 1. CSV 
df = pd.read_csv("jobrole_workload_summary.csv", index_col=0)

#  
sns.set(style="whitegrid")
plt.rcParams.update({"figure.figsize": (10,6)})

# ---------------------------------------------------
# 2. Total_Workload   
plt.figure()
df_sorted_total = df.sort_values("Total_Workload", ascending=False)
sns.barplot(x=df_sorted_total["Total_Workload"],
            y=df_sorted_total.index,
            palette="Blues_r")
plt.title("Total Workload by JobRole (Descending)")
plt.xlabel("Total Workload")
plt.ylabel("JobRole")
plt.tight_layout()
plt.show()

# ---------------------------------------------------
# 3. Extra_Cost   
plt.figure()
df_sorted_cost = df.sort_values("Extra_Cost", ascending=True)
sns.barplot(x=df_sorted_cost["Extra_Cost"],
            y=df_sorted_cost.index,
            palette="Greens")
plt.title("Extra Cost by JobRole (Ascending)")
plt.xlabel("Extra Cost")
plt.ylabel("JobRole")
plt.tight_layout()
plt.show()

# ---------------------------------------------------
# 4.      
# ( ,    )
for col in df.columns:
    print(f"\n=== {col}  ===")
    print(df[col].sort_values(ascending=False))
```

```python
import pandas as pd

# 1. CSV  
salary = pd.read_csv("department_jobrole_salary_summary.csv")
workload = pd.read_csv("department_jobrole_workload_summary.csv")

# workload  MultiIndex(Department, JobRole) reset_index 
if isinstance(workload.index, pd.MultiIndex):
    workload = workload.reset_index()

# 2. Department, JobRole  
# how='outer'        
merged = pd.merge(
    salary,
    workload,
    on=["Department", "JobRole"],
    how="outer"
)

# 3. CSV 
merged.to_csv("total.csv", index=False, encoding="utf-8-sig")

print("   : total.csv")
print(merged.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ---------------------------------------------------
# 1.  
df = pd.read_csv("total.csv", encoding="utf-8-sig")

# 2.   
df["__1"] = df["Total_Workload"] / df[""]
df["_"] = df["Extra_People_Needed"] * df[""]

# +    x 
df["_"] = df["Department"] + " - " + df["JobRole"]

# ---------------------------------------------------
sns.set(style="whitegrid")
plt.figure(figsize=(12,6))
# 3-1. 1   ()
order1 = df.sort_values("__1", ascending=False)
sns.barplot(data=order1, x="__1", y="_", palette="Blues_r")
plt.title("+ 1  ")
plt.xlabel("1  ")
plt.ylabel(" - ")
plt.tight_layout()
plt.show()

# ---------------------------------------------------
plt.figure(figsize=(12,6))
# 3-2.    
order2 = df.sort_values("Extra_People_Needed", ascending=False)
sns.barplot(data=order2, x="Extra_People_Needed", y="_", palette="Oranges_r")
plt.title("+    ")
plt.xlabel("   ")
plt.ylabel(" - ")
plt.tight_layout()
plt.show()

# ---------------------------------------------------
plt.figure(figsize=(12,6))
# 3-3.        
order3 = df.sort_values("_", ascending=False)
sns.barplot(data=order3, x="_", y="_", palette="Greens_r")
plt.title("+      ")
plt.xlabel("   ( )")
plt.ylabel(" - ")
plt.tight_layout()
plt.show()
```
